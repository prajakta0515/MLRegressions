```python
import numpy as np
import pandas as pd 
from matplotlib  import pyplot as plt
```

```python
dataset = pd.read_csv('Salary_Data.csv')
```

```python
dataset.head()
```

```python
X=dataset.iloc[:,:-1].values
```

```python
print(type(X))
```

```python
y = dataset.iloc[:,1].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, train_size=0.7, random_state= 120)
```

```python
from sklearn.linear_model import LinearRegression
model= LinearRegression()
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
X_pred=model.predict(X_train)
```

```python
from sklearn.metrics import r2_score
r2= r2_score(y_test,y_pred)
r2
```

```python
# visualising training  set results
plt.scatter(X_test,y_test, color='blue')
plt.plot(X_train,X_pred,color='green')
plt.xlabel('Yearsofexperience')
plt.ylabel('salary')
plt.show()
```

```python
X=customers [['Avg. Session Length','Time on App','Length of Membership']]
y=customers['Yearly Amount Spent']
```

```python
from sklearn.model_selection import train_test_split

# Ensure that this line is properly indented relative to the previous code
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression

mlmodel= LinearRegression()
mlmodel.fit(X_train,y_train)
```

```python
print(mlmodel.coef_[0],mlmodel.coef_[1],mlmodel.coef_[2])
```

```python
result=mlmodel.predict(X_test)
```

```python
from sklearn import metrics
print('R2 score:',metrics.r2_score(y_test,result))
print('mean square error:',metrics.mean_squared_error(y_test,result))
```

```python
plt.scatter(y_test,result)
plt.xlabel('Actuals')
plt.ylabel('Predicted')
```
